```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/refs/heads/main/Task%204/twitter_training.csv"
df = pd.read_csv(url)


df.columns = ["Tweet_ID", "Entity", "Sentiment", "Text"]


sentiment_counts = df["Sentiment"].value_counts()
print("\nSentiment counts:")
print(sentiment_counts)


plt.figure(figsize=(6,4))
sns.countplot(x="Sentiment", data=df, palette="viridis")
plt.title("Sentiment Distribution (Bar Chart)")
plt.show()


plt.figure(figsize=(6,6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("viridis", len(sentiment_counts)))
plt.title("Sentiment Distribution (Pie Chart)")
plt.show()


plt.figure(figsize=(12,6))
sns.countplot(x="Entity", hue="Sentiment", data=df, palette="viridis")
plt.title("Entity-wise Sentiment Analysis")
plt.xlabel("Entity (Brand/Topic)")
plt.ylabel("Number of Tweets")
plt.xticks(rotation=90)
plt.legend(title="Sentiment")
plt.show()
```
